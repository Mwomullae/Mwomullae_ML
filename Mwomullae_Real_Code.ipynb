# 결측지 해결
<hr>

### 결측치를 해당 Item의 평균 rating으로 채웠을 때와 0으로 채웠을 때의 성능 차이가 약9%가량 났음

```python
# 결측치를 0이 아닌 값들의 평균으로 채우는 함수
def fill_na_with_mean(dataset):
    for column in datset_col:
        non_zero_values = dataset[column][dataset[column] != 0]           # 0이 아닌 값들을 선택
        mean = non_zero_values.mean() if len(non_zero_values) > 0 else 0  # 0이 아닌 값들의 평균 계산
        dataset[column] = dataset[column].replace(0, mean)                # 0을 평균값으로 채우기


# 함수 호출하여 결측치를 채우기
fill_na_with_mean(dataset)

dataset.head()
```

```python
row_len = len(dataset["부산삼겹살 하단본점"])
```

```python
index = 0
count = 1
column_list = dataset.columns.to_list()
column_len = len(column_list)



for i in range(column_len):
    if (count == 1):
        user_data = pd.DataFrame({f"{count}_rating":dataset[column_list[index]].values})
        length = len(user_data[f'{count}_rating'])
        index = index + 1
    else:
        user_data[f"{count}_rating"] = dataset[column_list[index]].values
        index = index + 1

    for _ in range(length):
        user_data[f"{count}_restaurant_id"] = count
    count = count + 1
user_data
```

```python
count = 1
user_id = []

for i in range(column_len):
    for i in range(row_len):
        user_id.append(count)
        count = count + 1
    count = 1
```

```python
# user_matrix 생성
count = 1
ration_data = []
restaurant_id_data = []

for i in range(35):
    i = user_data[f"{count}_rating"].to_list()
    ration_data.extend(i)

    i = user_data[f"{count}_restaurant_id"].to_list()
    restaurant_id_data.extend(i)
    count = count + 1

user_data = pd.DataFrame({"user_id": user_id, "restaurant_id":restaurant_id_data, "rating": ration_data})
user_data
```

```python
# item_matrix 생성
count = 1
restaurant = []
restaurant_id = []

for i in datset_col:
    restaurant.append(i)
    restaurant_id.append(count)
    count = count + 1

item_data = pd.DataFrame({"restaurant_id":restaurant_id, "restaurant": restaurant})
item_data
```

```python
# merge: user_data & item_data
# merge: 공통 column[restaurant_id]
user_item_data = pd.merge(user_data, item_data, on="restaurant_id")
user_item_data
```

# 06. Model Train
---

```python
from sklearn.decomposition import TruncatedSVD
# SVD 모델 학습
num_factors = 5  # 잠재요인 개수 설정
svd = TruncatedSVD(n_components=num_factors)
svd.fit(user_restaurant_rating)
```

```python
# 잠재요인 행렬 생성
user_factors = svd.transform(user_restaurant_rating)
restaurant_factors = svd.components_
```

---
# 07. Performance Evaluation
---

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error

# 재구성된 행렬의 평점 예측 값 계산
reconstructed_ratings = np.dot(user_factors, restaurant_factors)

# 재구성 오차 계산 [원본 행렬과 재구성 행렬의 차이 계산 (오차 계산)]
mse = mean_squared_error(user_restaurant_rating.values, reconstructed_ratings)
mae = mean_absolute_error(user_restaurant_rating.values, reconstructed_ratings)

print(f"평균 제곱 오차 (MSE): {round(mse, 2)}")
print(f"평균 절대 오차 (MAE): {round(mae,2)}")
```

```python
# MSE 값을 0에서 1 사이의 범위로 정규화하여 정확도 계산
normalized_error = mse / (user_restaurant_rating.values.max() - user_restaurant_rating.values.min())
accuracy = 1 - normalized_error
print(f"정확도: {round(accuracy * 100,2)}%")
```
