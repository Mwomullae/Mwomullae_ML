```python
import pandas as pd

dataset = pd.read_csv("/content/drive/MyDrive/dataset/user_item_real_data.csv", encoding="cp949")
dataset.head() 
```

---
# 02. Data Setting
---

```python
dataset = dataset.drop("타임스탬프", axis=1)
dataset.head()
```

```python
dataset.info()
```

```python
datset_col = dataset.columns.to_list()
print(datset_col)
```

```python
row_len = len(dataset["부산삼겹살 하단본점"])
```

```python
index = 0
count = 1
column_list = dataset.columns.to_list()
column_len = len(column_list) 



for i in range(column_len):
    if (count == 1):
        user_data = pd.DataFrame({f"{count}_rating":dataset[column_list[index]].values})
        length = len(user_data[f'{count}_rating'])
        index = index + 1
    else:
        user_data[f"{count}_rating"] = dataset[column_list[index]].values
        index = index + 1

    for _ in range(length):
        user_data[f"{count}_restaurant_id"] = count
    count = count + 1
user_data
```

```python
count = 1
user_id = []

for i in range(column_len):
    for i in range(row_len):
        user_id.append(count)
        count = count + 1
    count = 1
```

```python
# user_matrix 생성
count = 1
ration_data = []
restaurant_id_data = []

for i in range(35):
    i = user_data[f"{count}_rating"].to_list()
    ration_data.extend(i)

    i = user_data[f"{count}_restaurant_id"].to_list()
    restaurant_id_data.extend(i)
    count = count + 1

user_data = pd.DataFrame({"user_id": user_id, "restaurant_id":restaurant_id_data, "rating": ration_data})
user_data
```

```python
# item_matrix 생성
count = 1
restaurant = []
restaurant_id = []

for i in datset_col:
    restaurant.append(i)
    restaurant_id.append(count)
    count = count + 1

item_data = pd.DataFrame({"restaurant_id":restaurant_id, "restaurant": restaurant})
item_data
```

---
# 03. Data Merge
---

```python
# merge: user_data & item_data
# merge: 공통 column[restaurant_id] 
user_item_data = pd.merge(user_data, item_data, on="restaurant_id")
user_item_data
```

```python
user_item_data.shape
```

---
# 04. Change Data
---

---
* row : user_id
* column : restaurant
* value : rating 
---

```python
user_restaurant_rating = user_item_data.pivot_table("rating", index="user_id", columns="restaurant").fillna(0)
```

# 06. Model Train
---

```python
from sklearn.decomposition import TruncatedSVD

# SVD 모델 학습
num_factors = 15  # 잠재요인 개수 설정
svd = TruncatedSVD(n_components=num_factors)
svd.fit(user_restaurant_rating)
```

```python
# 잠재요인 행렬 생성
user_factors = svd.transform(user_restaurant_rating)
restaurant_factors = svd.components_
```

---
# 07. Performance Evaluation
---

```python
import numpy as np
from sklearn.metrics import mean_squared_error

reconstructed_ratings = np.dot(user_factors, restaurant_factors)    # 재구성된 행렬의 평점 예측 값 계산
error = mean_squared_error(user_restaurant_rating.values, reconstructed_ratings)    # 원본 행렬과 재구성 행렬의 차이 계산 (오차 계산)
print(f"평점 예측 오차 (Mean Squared Error): {error}")
```

```python
# 사용자에게 음식점 추천
user_id = 1 # 추천을 받을 사용자의 ID 입력
user_idx = user_id - 1
user_ratings = np.dot(user_factors[user_idx, :], restaurant_factors)
sorted_restaurant_indices = np.argsort(user_ratings)[::-1][:5]
```

```python
# 추천된 음식점 출력
recommended_restaurants = user_restaurant_rating.columns[sorted_restaurant_indices]
print(f"[ User {user_id}에게 추천되는 음식점 ]")
print()
for restaurant in recommended_restaurants:
    print(restaurant)
```
